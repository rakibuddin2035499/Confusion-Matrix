# Seperate X, Y

```python
x = df.drop('DEATH_EVENT', axis='columns')
```

```python
x.head()
```

```python
y = df['DEATH_EVENT']
```

```python
y.head()
```

# Split the dataset into train and test

```python
xtrain, xtest, ytrain, ytest = tts(x,y,test_size=.30,random_state=1)
```

```python
xtrain.head()
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
clf = DecisionTreeClassifier() # creating object for decision tree
```

```python
clf.fit(xtrain, ytrain)
```

```python
from sklearn import tree
plt.figure(figsize=(20,20))
tree.plot_tree(clf, filled=True, rounded=True, feature_names=x.columns)
plt.savefig('tree.png', dpi=1200)
```

```python
pred = clf.predict(xtest)
```

```python
pred # predicted result
```

```python
np.array(ytest) # actual result
```

# Precision or Positive predictive Value(PPV)

```python
precision = tp/(tp+fp)
```

```python
precision
```

# Sensitivity,Recall,Hit Rate,Or True Positive Rate(TPR)

```python
tpr = tp/(tp+fn)
```

```python
tpr
```

# False Positive Rate (FPR) Or 1-Spcificity

```python
fpr = fp/(fp+tn)
```

```python
fpr
```

# F1 measure

```python
F1Measure=(precision+tpr)/2
```

```python
F1Measure
```

# Harmonic Mean,F1 Score

```python
F1=(2*precision*tpr)/(precision+tpr)
```

```python
F1
```

# Specificity,Selectivity or Negative Rate (TNR)

```python
spc = tn/(tn+fp)
```

```python
spc
```

# Threat Score (TS) or Critical Success Index(CSI)

```python
csi = tp/(tp+fn+fp)
```

```python
csi
```

# False Discovery Rate (FDR)

```python
fdr = fp / (tp + fp)
```

```python
fdr
```
